```python
from rouge_score import rouge_scorer
import numpy as np
from sentence_transformers import SentenceTransformer, util

# Load the pre-trained or fine-tuned model
model = SentenceTransformer('all-MiniLM-L6-v2')

# Input text (dialogue)
text = """Miranda: Hi S, could we cancel tomorrow's meeting?
Stephanie: again? what happened?
Miranda: I have to work :(
Stephanie: you work 20 hours a day!
Miranda: that's what young doctors do
Stephanie: they kill themselves working
Miranda: no they come close to death but survive and it makes them immortal
Stephanie: as long as you have an explanation :D"""

# Reference summary (ground truth)
reference_summary = """Miranda: that's what young doctors do. Stephanie: they kill themselves working. Miranda: no they come close to death but survive and it makes them immortal."""

# Split the text into sentences
sentences = text.split('\n')

# Encode the sentences using the model
sentence_embeddings = model.encode(sentences)

# Calculate the cosine similarity between each sentence embedding
similarity_matrix = np.zeros([len(sentences), len(sentences)])
for i in range(len(sentences)):
    for j in range(len(sentences)):
        if i != j:
            similarity_matrix[i][j] = util.cos_sim(sentence_embeddings[i], sentence_embeddings[j])

# Calculate sentence scores by summing the cosine similarity values
sentence_scores = np.sum(similarity_matrix, axis=1)

# Select the top 3 sentences with the highest scores
summary_sentences = []
for i in range(3):
    index = np.argmax(sentence_scores)
    summary_sentences.append(sentences[index].strip())
    sentence_scores[index] = -1  # To avoid selecting the same sentence again

# Concatenate the summary sentences to create the final summary
generated_summary = '. '.join(summary_sentences)

# ROUGE score calculation using rouge_scorer
scorer = rouge_scorer.RougeScorer(["rouge1", "rouge2", "rougeL"], use_stemmer=True)
scores = scorer.score(reference_summary, generated_summary)

# Print ROUGE scores
print("ROUGE Scores:")
print(f"ROUGE-1: {scores['rouge1']}")
print(f"ROUGE-2: {scores['rouge2']}")
print(f"ROUGE-L: {scores['rougeL']}")

# Accuracy calculation: measure how many sentences from the generated summary appear in the reference
generated_sentences = set(generated_summary.split('. '))
reference_sentences = set(reference_summary.split('. '))

# Calculate accuracy
correct = len(generated_sentences.intersection(reference_sentences))
accuracy = correct / len(generated_sentences) * 100

print(f"Accuracy: {accuracy:.2f}%")
```

```python
from sentence_transformers import SentenceTransformer, InputExample, losses
from torch.utils.data import DataLoader
from datasets import load_dataset
from sklearn.metrics import f1_score, precision_score, recall_score
import torch
from tqdm import tqdm

# Load the dataset
dataset = load_dataset('/content/drive/MyDrive/samsum')

# Convert the dataset into a list of InputExamples for training
train_examples = []
for dialog, summary in zip(dataset['train']['dialogue'], dataset['train']['summary']):
    train_examples.append(InputExample(texts=[dialog, summary]))

# Load the pre-trained SentenceTransformer model
model = SentenceTransformer('all-MiniLM-L6-v2')

# Create a DataLoader for training
train_dataloader = DataLoader(train_examples, shuffle=True, batch_size=16)

# Define the loss function
train_loss = losses.MultipleNegativesRankingLoss(model)

# Fine-tuning the model (without evaluation for simplicity here)
model.fit(
    train_objectives=[(train_dataloader, train_loss)],
    epochs=1,
    warmup_steps=100
)

# Save the fine-tuned model (optional)
model.save('/content/drive/MyDrive/fine_tuned_samsum_model')

# Evaluation function
def compute_metrics(model, dataset):
    dialogues = dataset['dialogue']
    summaries = dataset['summary']

    # Get embeddings for dialogues and summaries
    dialogue_embeddings = model.encode(dialogues, convert_to_tensor=True)
    summary_embeddings = model.encode(summaries, convert_to_tensor=True)

    # Calculate cosine similarity between dialogues and summaries
    cosine_similarities = torch.nn.functional.cosine_similarity(dialogue_embeddings, summary_embeddings)

    # Threshold cosine similarities to get predictions (adjust threshold as needed)
    threshold = 0.5  # You can experiment with this threshold
    predictions = (cosine_similarities >= threshold).int().tolist()

    # Ground truth for F1 calculation
    ground_truth = [1] * len(predictions)  # Assuming binary classification for matching

    # Compute F1 score, precision, recall
    f1 = f1_score(ground_truth, predictions)
    precision = precision_score(ground_truth, predictions)
    recall = recall_score(ground_truth, predictions)

    return f1, precision, recall

# Use the function to evaluate the model on the validation set (if available)
val_f1, val_precision, val_recall = compute_metrics(model, dataset['validation'])

print(f'Validation F1 Score: {val_f1}')
print(f'Validation Precision: {val_precision}')
print(f'Validation Recall: {val_recall}')
```

```python
from transformers import AutoTokenizer

# Load a pretrained English tokenizer (GPT-2 in this case)
tokenizer = AutoTokenizer.from_pretrained("gpt2")

# Ensure the tokenizer has a padding token
if tokenizer.pad_token is None:
    tokenizer.pad_token = tokenizer.eos_token  # Use eos_token as pad_token

# Preprocessing function
def preprocess_function(examples):
    return tokenizer(
        examples["dialogue"],
        truncation=True,
        padding="max_length",
        max_length=128
    )

# Tokenize train and validation datasets
tokenized_datasets = dataset.map(preprocess_function, batched=True)
```

```python
from transformers import DataCollatorForLanguageModeling, Trainer, TrainingArguments, AutoModelForCausalLM

# Load GPT-2 model
model = AutoModelForCausalLM.from_pretrained("gpt2")

# Data collator for causal language modeling
data_collator = DataCollatorForLanguageModeling(
    tokenizer=tokenizer,
    mlm=False,  # Causal language modeling
)

# Define training arguments
training_args = TrainingArguments(
    output_dir="./gpt2-samsum",
    overwrite_output_dir=True,
    num_train_epochs=1,
    per_device_train_batch_size=8,
    save_steps=500,
    save_total_limit=2,
    evaluation_strategy="steps",
    eval_steps=200,
    warmup_steps=200,
    logging_dir="./logs",
)

# Initialize Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_datasets["train"],
    eval_dataset=tokenized_datasets["validation"],
    data_collator=data_collator,
)

# Train the model
trainer.train()

# Save the model
trainer.save_model("./gpt2-samsum")
```

```python
from rouge_score import rouge_scorer
from transformers import AutoTokenizer, AutoModelForCausalLM

# Load GPT-2 model and tokenizer
tokenizer = AutoTokenizer.from_pretrained("gpt2")
model = AutoModelForCausalLM.from_pretrained("gpt2")

# Ensure the tokenizer has a padding token
if tokenizer.pad_token is None:
    tokenizer.pad_token = tokenizer.eos_token  # Use eos_token as pad_token

# Function to generate summaries
def generate_summary(input_text, max_length=510, temperature=0.7, top_p=0.9):
    """
    Generate a summary for the input text using GPT-2.

    Args:
        input_text (str): The input text to summarize.
        max_length (int): The maximum length of the generated summary.
        temperature (float): Sampling temperature; higher values make output more random.
        top_p (float): Top-p sampling threshold.

    Returns:
        str: The generated summary.
    """
    # Append "TL;DR" to the input text
    prompt = f"{input_text}\n\nTL;DR:"

    # Encode the prompt
    input_ids = tokenizer.encode(prompt, return_tensors="pt").to(model.device)  # Move to model's device

    # Generate output
    output = model.generate(
        input_ids,
        max_length=max_length,
        temperature=temperature,
        top_p=top_p,
        top_k=50,
        eos_token_id=tokenizer.eos_token_id,
        pad_token_id=tokenizer.pad_token_id,
    )

    # Decode and return the generated text
    generated_text = tokenizer.decode(output[0], skip_special_tokens=True)
    # Extract the summary after "TL;DR:"
    summary = generated_text.split("TL;DR:")[-1].strip()
    return summary

# Function to calculate ROUGE score
def calculate_rouge_score(reference_summary, generated_summary):
    """
    Calculate ROUGE score between reference and generated summary.

    Args:
        reference_summary (str): The reference summary (ground truth).
        generated_summary (str): The generated summary.

    Returns:
        dict: ROUGE score metrics (ROUGE-1, ROUGE-2, ROUGE-L).
    """
    # Initialize the ROUGE scorer
    scorer = rouge_scorer.RougeScorer(
        ["rouge1", "rouge2", "rougeL"], use_stemmer=True
    )

    # Compute ROUGE scores
    scores = scorer.score(reference_summary, generated_summary)

    return scores

# Example
input_text = "John and Sarah had a meeting scheduled at 10 AM to discuss the quarterly report. However, John was running late due to traffic, so he texted Sarah to postpone the meeting to 10:30 AM."
reference_summary = "John was late, so the meeting with Sarah was postponed to 10:30 AM."

# Generate the summary
generated_summary = generate_summary(input_text)

# Calculate ROUGE score
rouge_scores = calculate_rouge_score(reference_summary, generated_summary)

# Print results

print("\nROUGE Scores:")
print(rouge_scores)
```
